```python
UCLA=[31,40,37,27,31,33,32,35,36,40,44]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing, model_selection
plt.figure(figsize=(9, 7))
Years=list(range(2013,2024))
plt.scatter(Years,UCLA,color="green",label="UCLA")
plt.legend(loc="upper left")
plt.xticks(Years,fontsize=7)
plt.ylim(50,15)
plt.xlabel('Years')
plt.ylabel('Ranking')
plt.title("QS_Ranking")
```

# SVR

```python
from sklearn.svm import SVR
plt.figure(figsize=(9, 7))
svr_rbf = SVR(kernel="rbf", C=10000, gamma=0.001, epsilon=0.1)
X=pd.DataFrame(Years)
y_pred=svr_rbf.fit(X,UCLA).predict(X)
plt.scatter(Years,UCLA,color="red",label="UCLA")
plt.legend(loc="upper left")
plt.xticks(Years,fontsize=8)
plt.ylim(50,20)
plt.xlabel('Years')
plt.ylabel('Ranking')
plt.title("QS_Ranking")
plt.plot(Years,y_pred, color='blue',linewidth=2,linestyle='--')
```

```python
from sklearn.svm import SVR
x_data=np.linspace(-5,5,200)
y_data=x_data**3+np.random.normal(0,20,x_data.shape)
plt.figure(figsize=(9, 7))
svr_rbf = SVR(kernel="rbf", C=10000, gamma=0.01, epsilon=0.01)
y_pred1=svr_rbf.fit(pd.DataFrame(x_data),list(y_data)).predict(pd.DataFrame(x_data))
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred1, color='red',linewidth=3,linestyle='-')
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_data, y_pred1)
print(r2)
```

# Poly_Regression

```python
X = np.array(list(range(2013,2024))).reshape(-1,1)
poly3 = preprocessing.PolynomialFeatures(3)
poly3X = poly3.fit_transform(X)
```

```python
poly3reg = linear_model.LinearRegression(fit_intercept = False)
poly3reg.fit(poly3X, UCLA)
print(poly3reg.score(poly3X,UCLA))
```

```python
x_predict = np.arange(2013, 2024).reshape(-1,1)
```

```python
poly3_X_new = poly3.fit_transform(x_predict)
poly3_y_hat = poly3reg.predict(poly3_X_new)
```

```python
plt.figure(figsize=(9, 7))
plt.scatter(Years,UCLA,color="green",label="UCLA")
plt.plot(Years,poly3_y_hat,color='blue',linewidth=2,linestyle='--')
plt.legend(loc="upper left")
plt.xticks(Years,fontsize=7)
plt.ylim(50,20)
plt.xlabel('Years')
plt.ylabel('Ranking')
plt.title("QS_Ranking")
```
